```python
# by this method we can handle all kind of exception but user may not understand this
try:
    a=int(input("Enter first number: "))
    b=int(input("Enter second number: "))
    c=a/b
    print(c)
except Exception as e:
    print(e)
```

```python
# we can customize this and can display something user understandable
try:
    a=int(input("Enter first number: "))
    b=int(input("Enter second number: "))
    c=a/b
    print(c)
except ZeroDivisionError:
    print("Please insert number grater than zero")
except Exception as e:
    print(e)
```

```python
# we can handle other exception as well
try:
    a=int(input("Enter first number: "))
    b=int(input("Enter second number: "))
    c=a/d
    print(c)
except ZeroDivisionError:
    print("Please insert number grater than zero")
except NameError:
    print("Some varaible may not defined/ declared")
except Exception as e:
    print(e)
```

```python
# using fially block, finally block will always executed, it will help us to write important code in this block. If some exception happpens in other block it will still execute finally block
try:
    a=int(input("Enter first number: "))
    b=int(input("Enter second number: "))
    c=a/d
    print(c)
except ZeroDivisionError:
    print("Please insert number grater than zero")
except NameError:
    print("Some varaible may not defined/ declared")
except Exception as e:
    print(e)
finally:
    print("Successfully executed")
```

```python
# Another example of finally block where all code executed perfectly
try:
    a=int(input("Enter first number: "))
    b=int(input("Enter second number: "))
    c=a/b
    print(c)
except ZeroDivisionError:
    print("Please insert number grater than zero")
except NameError:
    print("Some varaible may not defined/ declared")
except Exception as e:
    print(e)
finally:
    print("Successfully executed")
```

```python
# we can also use else block where we can perfrom operations and display the result
try:
    a=int(input("Enter first number: "))
    b=int(input("Enter second number: "))
except ZeroDivisionError:
    print("Please insert number grater than zero")
except NameError:
    print("Some varaible may not defined/ declared")
except Exception as e:
    print(e)
else:
    c=a/b
    print(c)
finally:
    print("Successfully executed")
```
